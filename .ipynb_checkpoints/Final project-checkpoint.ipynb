```python
data = pd.read_csv(r"C:\Users\valtt\Desktop\Spring 23\02806 Social data analysis and visualization\FinalProject\Pysakointivirheet_2014_17.csv", encoding='latin-1',converters={'Postinumero': str}) # Load the data
data
```

```python
data = data.reset_index()
data = data.rename(columns={"index": "Tickets", "Virheen tekokuukausi" : "Month", "Virheen tekovuosi" : "Year", "Osoite" : "Address", "Virhemaksun vaihe" : "Fine type", "Virheen pääluokka / pääsyy" : "Type of offence", "Virheen kirjaaja" : "Fine issuer", "Postinumero": "Postal code", "Postitoimipaikka": "Postal district", "Alue": "Area", "Kunta": "Municipality", "Kunta_nro": "Municipality number"  })
data
```

```python
data['Fine type'].unique()
```

```python
# create empty column 'Month number'
data['Month number'] = data.apply(lambda _: '', axis=1)
```

```python
# creating month number
data.loc[data.Month == 'Tammikuu', 'Month number'] = "1"
data.loc[data.Month == 'Helmikuu', 'Month number'] = "2"
data.loc[data.Month == 'Maaliskuu', 'Month number'] = "3"
data.loc[data.Month == 'Huhtikuu', 'Month number'] = "4"
data.loc[data.Month == 'Toukokuu', 'Month number'] = "5"
data.loc[data.Month == 'Kesäkuu', 'Month number'] = "6"
data.loc[data.Month == 'Heinäkuu', 'Month number'] = "7"
data.loc[data.Month == 'Elokuu', 'Month number'] = "8"
data.loc[data.Month == 'Syyskuu', 'Month number'] = "9"
data.loc[data.Month == 'Lokakuu', 'Month number'] = "10"
data.loc[data.Month == 'Marraskuu', 'Month number'] = "11"
data.loc[data.Month == 'Joulukuu', 'Month number'] = "12"
data
```
